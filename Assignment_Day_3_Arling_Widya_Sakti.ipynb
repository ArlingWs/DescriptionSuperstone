```python
import pandas as pd

```

```python
#Definiskan Kolom
Nama = ['Alex','Budi','Chandra','David','Edo','Farah','Gilang','Harry','Intan','Joko']
Tempat_lahir = ['Magelang','Bantul','Purwokerto','Jakarta','Palembang','Papua','Yogyakarta','Surabaya','Bali','Cirebon']
Tanggal_lahir = ['1994-12-15','1997-05-19','1991-11-28','1999-03-03','1996-06-10','1992-02-25','1998-08-09','2000-10-08','1993-09-17','1989-07-23']
Tanggal_terakhir_berbelanja = ['2020-12-15','2020-11-18','2020-03-11','2020-01-03','2020-05-08','2020-02-11','2020-05-10','2020-03-27','2020-04-28','2020-17-16']
amount_belanja_per_bulan = [189000,79000,145000,232000,94000,135000,186000,83000,130000,115000]
Total_transaksi_per_bulan = [7,2,11,19,4,9,16,3,12,6]

#Membuat Dictionary 'Nama Kolom' : List_data
data_user ={ 'Nama':Nama,
            'Tempat_lahir': Tempat_lahir,
            'Tanggal_lahir': Tanggal_lahir,
            'Tanggal_terakhir_berbelanja': Tanggal_terakhir_berbelanja,
            'amount_belanja_per_bulan': amount_belanja_per_bulan,
            'Total_transaksi_per_bulan': Total_transaksi_per_bulan
}
```

```python
##pd.DataFrame(Dictionary)
df_data_user = pd.DataFrame(data_user)
```
